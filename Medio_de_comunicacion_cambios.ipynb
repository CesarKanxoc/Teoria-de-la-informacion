```python
import random
import base64
import re
import math

# Función para calcular la entropía de una lista de bytes
def calculate_entropy(data):
    if not data:
        return 0.0

    entropy = 0
    size = len(data)
    for byte in set(data):
        probability = float(data.count(byte)) / size
        entropy -= probability * math.log2(probability)
    return entropy

def encode_book(file_path):
    try:
        with open(file_path, "rb") as archivo:
            file_content = archivo.read()
            encoded_content = base64.b64encode(file_content)

        packet_size = 1024  # Tamaño del paquete en bytes
        packets = [encoded_content[i:i + packet_size] for i in range(0, len(encoded_content), packet_size)]
        return packets
    except FileNotFoundError:
        print(f"El archivo '{file_path}' no se encontró.")
        return []

def apply_wifi_noise(packet):
    # Simular la pérdida de paquetes con una probabilidad del 10%
    if random.random() < 0.1:
        return None
    
    # Simular errores de bit con una probabilidad del 5%
    error_prob = 0.05
    noisy_packet = bytearray(packet)
    for i in range(len(noisy_packet)):
        if random.random() < error_prob:
            noisy_packet[i] = random.randint(0, 255)
    
    return bytes(noisy_packet)

# Función para simular un canal de WiFi
def simulate_wifi_channel(packets):
    noisy_packets = []
    for packet in packets:
        noisy_packet = apply_wifi_noise(packet)
        if noisy_packet is not None:
            noisy_packets.append(noisy_packet)
    return noisy_packets

# Función para decodificar los paquetes (sin cambios en esta versión)
def decode_packets(noisy_packets):
    return noisy_packets

# Función para reconstruir el libro con los cambios solicitados
def reconstruct_book(noisy_packets):
    combined_data = b''.join(noisy_packets)
    book_content = base64.b64decode(combined_data)
    return book_content

# Nombre del archivo de entrada
input_file_path = "war_and_peace.txt"
# Codificar el libro en base64
book = encode_book(input_file_path)

# Simular la transmisión a través de un canal WiFi
book_env = simulate_wifi_channel(book)

# Decodificar los paquetes recibidos (sin cambios)
book_decode = decode_packets(book_env)

# Reconstruir el libro con los cambios
book_reconst = reconstruct_book(book_decode)

# Calcular la entropía del ruido en el canal WiFi
entropy = calculate_entropy(book_env)
print(f"Entropía del ruido en el canal WiFi: {entropy}")

# Nombre del archivo de salida
output_file_path = "war_and_peace2.txt"
# Guardar el libro reconstruido en un nuevo archivo
with open(output_file_path, 'wb') as f:
    f.write(book_reconst)

```
